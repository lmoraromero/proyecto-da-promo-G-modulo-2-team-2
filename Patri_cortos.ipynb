```python
def call_movies_api(page, genre, year):
    url = "https://moviesdatabase.p.rapidapi.com/titles"
    querystring = {"genre": genre, "titleType": "short", "year": year, "limit": "50", "page": page}
    headers = {
        "X-RapidAPI-Key": "09af841796mshace1fa6f270a9c5p1bed4ajsnbbf39674fe5f",
        "X-RapidAPI-Host": "moviesdatabase.p.rapidapi.com"
    }
    response = requests.get(url, headers=headers, params=querystring)
    print(f"He descargado del genero {genre} del año {year} para la página {page}")
    return response.json()

def get_page(response):
    if "next" in response and response["next"] is not None:
        next_url = response["next"]
        regex = r"(page=)(.*?)(&|$)"
        return re.search(regex, next_url).group(2)
    else:
        return None

genres = ["Drama", "Comedy", "Action"]
lista_pelis = []

for genre in genres:
    for year in tqdm(range(2021, 2022)):
        page = 1
        while page:
            response = call_movies_api(page, genre, year)
            if response:
                for pelicula in response['results']:
                    tipo = pelicula['titleType']['text']
                    titulo = pelicula['titleText']['text']
                    estreno_a = pelicula['releaseYear']['year']
                    estreno_m = pelicula['releaseDate']['month'] if pelicula['releaseDate'] else None
                    id_pel = pelicula['id']
                    tupla_pelis = (tipo, titulo, estreno_a, estreno_m, id_pel)
                    lista_pelis.append(tupla_pelis)
                page = get_page(response)
            else:
                break

print(lista_pelis)








```

```python
df_pelis = pd.DataFrame(lista_pelis, columns=['Tipo', 'Título', 'Año de estreno', 'Mes de estreno', 'ID'])
```
