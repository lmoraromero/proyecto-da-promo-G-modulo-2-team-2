Fase 4: Extracción de Tablas de los Premios Oscar con Beautiful Soup   
En esta fase, trabajarán con la biblioteca Beautiful Soup para extraer información relevante de las tablas de los premios Oscar desde 2000. La tabla la encontrarás en el este link. La información que deberás sacar es:   
    - Fecha de la ceremonia.   
    - Mejor película.   
    - Mejor director.   
    - Mejor actor.   
    - Mejor actriz.   
    
    NOTA: La información de la API deberá ser almacenada en una lista de tuplas. Cada tupla corresponderá a una película. Siguiendo el siguiente ejemplo:

```python
#Hacer la llamada 

url_oscars = "https://es.wikipedia.org/wiki/Premios_%C3%93scar"
res_oscars = requests.get(url_oscars)
sopa_oscars = BeautifulSoup(res_oscars.content, "html.parser")
```

```python
tablas = sopa_oscars.find_all('table', {'class': 'wikitable'})
```

```python
tabla = tablas[0]
```

```python
filas = tabla.find_all('tr')[72:]
```

```python
len(filas)
```

```python
filas[0]
```

```python
dicc_oscars = {'ceremonia':[], 'pelicula':[], 'director':[], 'actor':[], 'actriz':[]}

for fila in filas:
    celdas = fila.find_all('td')
    dicc_oscars['ceremonia'].append(celdas[1].text.strip())
    dicc_oscars['pelicula'].append(celdas[2].text.strip())
    dicc_oscars['director'].append(celdas[3].text.strip())
    dicc_oscars['actor'].append(celdas[4].text.strip())
    dicc_oscars['actriz'].append(celdas[5].text.strip())

lista_oscars = list(zip(dicc_oscars['ceremonia'], dicc_oscars['pelicula'], dicc_oscars['director'], dicc_oscars['actor'], dicc_oscars['actriz']))
```

```python
df_oscars = pd.DataFrame(lista_oscars, columns=['Fecha ceremonia', 'Pelicula', 'Director', 'Actor', 'Actriz'])
```
