```python
diccionario_jug = {'nombres':[], 'apellidos':[], 'equipos':[]}

for pag in range(1, 9):
    url_jug = f'https://www.balldontlie.io/api/v1/players?page={pag}'
    jugadores = requests.get(url_jug) #requests de la url dinámica
    jugadores_json = jugadores.json() #json de cada llamada

    for jug in jugadores_json['data']: #itera por la key data
        nombre = jug['first_name']
        apellido = jug['last_name']
        equipo = jug['team']['full_name']

        diccionario_jug['nombres'].append(nombre) #accedo a la clave que es una lista vacía. por eso puedo hacer el el append
        diccionario_jug['apellidos'].append(apellido)
        diccionario_jug['equipos'].append(equipo)
```
