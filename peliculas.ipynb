```python
lista_pelis = []

for pagina in range(1, 6):
    print(pagina)
    url = "https://moviesdatabase.p.rapidapi.com/titles"

    querystring = {"genre":"Drama","titleType":"movie","year":"2007","page":pagina,"limit":"50"}

    headers = {
	"X-RapidAPI-Key": "600295e169msh9cf6b785b5b61cep134f8bjsn9aa642a83b89",
	"X-RapidAPI-Host": "moviesdatabase.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)
    response_json = response.json()

    for pelicula in response_json['results']:
        '''print(pelicula['titleType']['text'])
        print(pelicula['titleText']['text'])
        print(pelicula['releaseYear']['year'])
        print(pelicula['releaseDate']['month'])
        print(pelicula['id'])'''
        tipo = pelicula['titleType']['text']
        titulo = pelicula['titleText']['text']
        estreno_a = pelicula['releaseYear']['year']
        estreno_m = pelicula['releaseDate']['month'] if pelicula['releaseDate'] else None
        id_pel = pelicula['id']

        tupla_pelis = (tipo, titulo, estreno_a, estreno_m, id_pel)
        lista_pelis.append(tupla_pelis)
print(lista_pelis)
```

```python
columna_names = ['Tipo', 'Titulo', 'Año de estreno', 'Mes de estreno', 'ID']
```

```python
df_pelis = pd.DataFrame(lista_pelis, columns=columna_names)
```
